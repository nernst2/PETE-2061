```python
#Question 2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3

%matplotlib inline

fileName = 'Volve production data (1).xlsx'
xl = pd.ExcelFile(fileName)
dailyDF = xl.parse('Daily Production Data')

conn = sqlite3.connect('VolveProduction.db')
cur = conn.cursor()

titleFontSize = 18
axisLabelFontSize = 15
axisNumFontSize = 13

dailyDF.to_sql("DailyRates", conn, if_exists="append", index = False)

```

```python
wellBore = (5351, 5599, 5693, 5769, 7078, 7289, 7405)

for n in wellBore:
    testWellDF = pd.read_sql_query(f"SELECT DP_CHOKE_SIZE, BORE_OIL_VOL FROM DailyRates WHERE NPD_WELL_BORE_CODE={n};", conn)
    
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    ax1.plot(testWellDF['BORE_OIL_VOL'], color = "green", ls = 'None', marker = 'o', markersize = 2,)
    ax2.plot(testWellDF['DP_CHOKE_SIZE'], color = "blue", ls = 'None', marker = 'o', markersize = 2,)
    ax1.set_xlabel('Time, Days')
    ax1.set_ylabel('Cumulative Oil Production, Mbbls', color='g')
    ax2.set_ylabel('Choke Size', color='b')
    plt.show()
#control-shift-esc to see task manager 
```
