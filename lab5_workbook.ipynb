2. **Using the example in "lab5_Probability.ipynb", write a Python code to compute the probabililities of all possible total values that can be obtained from rolling a fair die three times**

```python
import itertools
import pandas as pd
import numpy as np


numberList = list(range(1,7))
resultList = list(itertools.product(numberList,repeat=3)) #rolling the die 3 times/rolling 3 dice  
print(f"Total number of outcomes for rolling three dice = {len(resultList)}")


resultArray = np.array(resultList)
sumOutcomes = np.sum(resultArray, axis=1)

df1 = pd.DataFrame({"outcomes":resultList,"sumOutcomes":sumOutcomes})
df1
```

```python
by_sumOutcomes = df1.groupby("sumOutcomes")

outcomeDF = by_sumOutcomes.count()
outcomes = outcomeDF.values[:,0]
probability = outcomes / np.sum(outcomes)
newDF = pd.DataFrame({"sum":list(range(3,19)), "outcomes":outcomes, "probability":probability})
newDF
```

```python
n=int(input())
numberList = list(range(1,7))
resultList = list(itertools.product(numberList,repeat=n)) #rolling the die n times/rolling n dice  
print(f"Total number of outcomes for rolling three dice = {len(resultList)}")

resultArray = np.array(resultList)
sumOutcomes = np.sum(resultArray, axis=1)

df1 = pd.DataFrame({"outcomes":resultList,"sumOutcomes":sumOutcomes})
df1
```

```python
by_sumOutcomes = df1.groupby("sumOutcomes")

outcomeDF = by_sumOutcomes.count()
outcomes = outcomeDF.values[:,0]
probability = outcomes / np.sum(outcomes)
newDF = pd.DataFrame({"sum":list(range(n,(6 * n)+1)), "outcomes":outcomes, "probability":probability})
newDF
```
